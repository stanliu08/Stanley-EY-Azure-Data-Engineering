#### Answer to #4

```python
# pretty sure this counts something like (rock), (rock, pop), and (pop) as 3 unique genres instead of 2
genre_list = df["genre"].nunique()
genre_list
```

```python
# splits all genres by comma before putting it into a large list
genre_split = df["genre"].str.split(",")
all_genres = [g.strip() for list in genre_split for g in list]
all_genres
# turn the long list into a set so we only have the unique genres
set(all_genres)
print(len(set(all_genres)))
# we have 15 unique genres
```

#### Answer to #6

```python
# does a group by for artists by their popularity
# uses .sum() to get the total, and sort valeus from greatest to least
pop_artist = df.groupby("artist")["popularity"].sum().sort_values(ascending=False)
# Rihanna is the most popular artist
```

#### Answer to #7

```python
# gets the index of where the longest song is
# idxmax() returns index of where the str len is highest
longest_song_index = df["song"].str.len().idxmax()

row_longest_song = df.loc[longest_song_index]
print(row_longest_song)
```

#### Answer to #8

```python
# gets the sum of the duration_ms column
all_songs_length = df["duration_ms"].sum()
print(all_songs_length)
# total duration of all songs is 457496249 ms
```
