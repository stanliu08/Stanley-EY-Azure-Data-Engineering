```python
# Collections

# List aka Dynamic Array
# There is no static array implementation in the standard libary
# Can use Numpy
drinks = ['Water', "Pepsi", 'Coke', 'Juice']
```

```python
# Tuple
# Maintains insertion order with an index.
# Allows for duplicates
# Immutable - Cannot be changed after instantiation, but you can reassign the name to a different value.
drinks = ('Water', "Pepsi", 'Coke', 'Juice')
drinks = ('Water', 'Beer', 'Water')
print(drinks)
```

```python
# Range
# Creates a sequence of numbers
# range(start, stop, step)
# Useful in for loops
my_list = []
for i in range(10):
  my_list.append(i)
print(my_list)

my_list = [i + 10 for i in range(0, 100, 10)]
print(my_list)
```

```python
# Sets
# Unique lists
# Set items are unordered, unchangeable, and unique.
# You cannot refer to the items by key.

my_set = {"apple", "banana", "cherry", "apple"}

print(my_set)

# If you need to get the length of any iterable object use len()
print(len(my_set))

# Sets can contain mixed data types.
mixed_set = {"apple", 3, True}
print(mixed_set)

# You cannot change items in a set, but they can be added or removed.
mixed_set.remove('apple')
print(mixed_set)
```

```python
# Dictionaries
# Store labeled data or key-value pairs
# Dictionaries are orders after version 3.7
# Items in a dictionary are mutable but do not allow duplicates.

employee = {
    'name': 'Jordan Miller',
    'position': 'Architect',
    'start_date': "04-11-2022",
    'certifications': 18
}

# We can index into dictionaries with the keys.
# Or we can use the get method.
name = employee['name']
print(name)
position = employee.get('position')
print(position)

# If we want all of the keys as a list
employee_keys = employee.keys()

# Similarly...
employee_values = employee.values()

# Lets say we need to operate on both...
employee_items = employee.items()
for k, v in employee_items:
  print(f"{k}: {v}")

# If we need to check if a key exists
# We cannot do this...
# if employee['key']
# We should do this instead
if 'position' in employee:
  print('The employee has a position.')


# If we need to update an item in a dictionary we have two options.
employee['certifications'] = 4
employee.update({'certifications': 5})

# A few methods are also available for removing items.
# For a single item
employee.pop('certifications')
del employee['position']

# To empty a dictionary
employee.clear()

# Del also works for deleting the entire dictionary
del employee

# Dictionaries can also be nested
myfamily = {
  "child1" : {
    "name" : "Emily",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
```

```python
# Boolean
# True or False
# Usually the result of a comparison

print(9 > 4)
print(11 == 11)
print(8 <= 4)

# The bool() function will perform a null check
# Pythons default null type is None
# The None keyword initializes an object of NoneType
# None is not the same as 0, False, or an empty string.
# Other libraries implement their own nulls like np.NaN (numpy Not a Number)

print(bool('hello'))
print(bool(None))

# Please don't ever do this.
condition = (5 > 4)
if condition == True:
  print('It\'s true.')

# This is much more Pythonic.
if condition:
  print("It's true.")
```

```python
# Casting Data Types

# int() creates an int from a float (by truncating) or a string
x = int(1)
y = int(2.8) # y = 2
z = int("3")

# float() constructs a float from an int, float literal, or string literal
x = float(1)
y = float(3.14)
z = float(3) # z = 3.0

# str() creates a string from the input
y = str(2)
z = str(3.0) 
```
