```python
# Date and times are maintained through the datetime module
import datetime

# Datetimes are of the format: Year, month, day, hour, minute, second, microsecond
# Two functions we will use the most are datetime.datetime.now()
# and datetime.date.today()

print(datetime.datetime.now())
print(datetime.date.today())
```

```python
# We want a standardized format to store datetimes, but we typically want
# to specify a format when we read them. We can use a couple of functions to
# help us with that. strftime and strptime (string format and parse time)
# For help with format strings https://www.w3schools.com/python/python_datetime.asp

now = datetime.datetime.now()

print(now.strftime('%A'))
print(now.strftime('%B %d, %Y'))
print(now.strftime('%d-%m-%y'))
print(now.strftime('%H:%M:%S'))
```

```python
# We also frequently need to do the opposite
# We are given date in a string format and need to extract a date.
string_date = '06-15-2022'
date = datetime.datetime.strptime(string_date, "%m-%d-%Y")
print(date)
```

```python
# Unix Time
# Number of seconds since Jan 1, 1970 UTC
# 32 bit int overflow on Jan 19, 2038
import time
print(time.time)
```
