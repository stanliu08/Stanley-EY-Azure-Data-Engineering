```python
# Modifying Strings

# Changing the casing
# Return the uppercase string with upper()
b = 'This is a string!'
print(b.upper())

# Or we can go to lowercase with lower()
print(b.lower())

# We can also do title casing
print(b.title())
```

```python
# If we want to replace something in the string.
print(b.replace('string', 'sentence'))
```

```python
# Not all of our text data is going to be very neat.
# Sometimes we need to clean it up a little bit.
# We can use split() and strip() to help with that.
c = "  This is a messy string.   "
print(c.strip())

# We can also specify characters to remove
d = ",,,,,rrttgg.....banana....rrr"
print(d.strip(",.grt"))

# The split function lets us separate a string into a list of words
# This is very a very common operation with data engineering.
words = c.split(' ')
print(words)

# Usually we combine it with strip
words = c.strip(" .").split(' ')
print(words)
```

```python
# In other languages we have to escape characters.
# Python gives us a few options to get around this.

# If I have a single quote in my string, I can wrap the string in double quotes.
text = "There's no issue here."
print(text)

# And vice versa.
text = 'How about "this"?'
print(text)

# However, we can always escape characters if we need to.
text = 'There\'s no issue here either.'
```

```python
# We may also want to construct dynamic strings to be customized.

first_name = 'Lee'
last_name = 'Speers'
text = "My name is {0} {1}"
print(text.format(first_name, last_name))

# The f-string lets us do this more pythonically.
# Start the string with an f before the first quote and then you can add the variable names in line.
print(f'My name is {first_name} {last_name}.')

# There is also an r-string for raw string.
# This makes the string a string literal.
# All of the characters in the string will be calculated exactly as is.
print(r'\\//\\//')
print('\\//\\//')
```

```python
# String Encoding
# Computers can't store letters and symbols in memory.
# It has to convert the human-readable characters into bytes.
# When text is saved to disk it is stored as a series of bytes with a schema.
# There are several different ways to encode the text we see into bytes.
# Python uses ASCII encodings by default (1-128)
# But there are other options
print('\u265E')

# The \u makes it a unicode character. In this case a black knight.
# If you have ever wondered why some url strings have funky characters, it is most likely the utf-8 encoding.
import urllib.parse
url = 'www.example.com?var=This%20is%20a%20simple%20%26%20short%20test.'
print(urllib.parse.unquote(url))

# This becomes increasingly important when dealing with APIs as query parameters are often encoded.
# Text files are also stored as bytes, and thus typically have an encoding schema.
# We will talk about this in an upcoming lecture.
```
