**EXERCISE 2**

Write a python program that asks the user to input ten positive integers, and then, prints out the largest odd, positive integer that was entered.  If no odd, positive integer was entered, then the program should print out a message indicating that this was the case.

````python
# NOTES/LEARNED:
# split() splits the input string into a list of strings based on whitespace
# map(int, input()) is converting the strings to integers "2" -> 2
# just using list(input("").split()) also works but it will be string characters instead of ints
# just realized we could make a loop that asks for each input until we have 10 ints, probably better for UX as well
ten_nums = list(map(int, input("Please enter 10 positive integers (separated with spaces): ").split()))


# Checks if exactly 10 ints were entered
if(len(ten_nums) != 10):
    print("Error: You must enter exactly 10 integers.")
else:
    # iterates to find largest odd, if none found then largest_odd remains 0
    largest_odd = 0
    for num in ten_nums:
        if(num > largest_odd and num % 2 != 0):
            largest_odd = num
    
    # prints the result
    if(largest_odd == 0):
        print("No odd numbers were entered.")
    else:
        print(f"The largest odd number is: {largest_odd}")
````

Design your function so that it supports both of these use cases:

* It can be called in a way that returns **either** just the list or just the dictionary, based on a function parameter.

Document via comments in the function on how the caller should use this parameter to control what the function returns.

````python
# NOTES: 
# ISSUES: (might fix if I have time)
# 1. Only counts uppercase letters, lowercase letters are ignored
# 2. Input validation, make sure it takes a string and only consists of the nucleotides
# Found an example online where you can add a return type in parameter instead of function overloadding

# TO USE: pass in only str to get both the list and dict, past in "list" or "dict" as second parameter to get only that type
def analyze_dna(str, return_type="both"):
    DNAfragment = str
    print(f"Length of entered string: {len(DNAfragment)}")
    
    # initializes variables, removed individual counters for A, T, C, and G since we're using a dict
    count = 0
    
    # list and dictionary to be used
    baseList = []
    baseDict = {'A':0, 'T':0, 'C':0, 'G':0}
    
    # iterates through the DNA strand and adds count to each matching nucleotide
    while count < len(DNAfragment):
        if DNAfragment[count] == 'A':
            baseList.append('A')
            baseDict['A'] += 1
        if DNAfragment[count] == 'T':
            baseList.append('T')
            baseDict['T'] += 1
        if DNAfragment[count] == 'C':
            baseList.append('C')
            baseDict['C'] += 1
        if DNAfragment[count] == 'G':
            baseList.append('G')
            baseDict['G'] += 1
        count += 1
        
    # print(f"Number of A nucleotides: {baseDict['A']}")
    
    # if statements to see what return type
    if return_type == "list":
        return baseList
    elif return_type == "dict":
        return baseDict
    else:
        return baseList, baseDict
        
# Should add input validation here?
str = input("Enter a DNA fragment: ")

# TO USE the analyze_dna(): pass in only str to get both the list and dict, past in "list" or "dict" as 
# second parameter to get only that type

# Example that returns both list and dict
dnalist1, dnadict1 = analyze_dna(str)
print(f"Returns both list and dict. \nThe DNA list is: {dnalist1},\nThe DNA dictionary is: {dnadict1}\n")

# Example that returns only list
dnalist2 = analyze_dna(str, "list")
print(f"Returns only the list. \nThe DNA list is: {dnalist2}\n")

# Example that returns only dict
dictlist2 = analyze_dna(str, "dict")
print(f"Returns only the dict. \nThe DNA dictionary is: {dictlist2}\n")

````

Write a function called *recursiveFind* that takes a list of integers as input and "walks down the list" recursively, from the first element of the list to the last element of the list, searching for the integer value assigned to a variable called *find_target* (*find_target* is a global variable that is assigned a value outside *recursiveFind*, but that is also used inside *recursiveFind* for comparison tests; see below **HINTS**).  Your program must print out to the command console the number of times that *find_target* is discovered in the list, like this:

        Total occurrences of find_target in list:  2.

Furthermore, your program **MUST NOT ALTER** the *original* input list!  Remember: Lists are **MUTABLE**! Thus, altering the original input list inside *recursiveFind* (for example, by "popping" it; see below **HINTS**) may get you into trouble becauses the altered list will then be passed into each subsequent, recursive call to the function *recursiveFind*.

Your program must be able to handle input lists of arbitrary length, including the empty list.
- If an empty list was input to *recursiveFind*, then your function *recursiveFind* should return the value 0 immediately.

**NOTE**:  This exercise **MUST BE SOLVED RECURSIVELY**.  This means you can't use something like the list method **find()**, or use iteration (such as a FOR or WHILE loop), or any means other than recursion.

````python
# NOTES:
# Global variable, apparently you don't need a keyword to declare it
# If it's outside func or class, it becomes a global variable
# Learned that instead of creating a new parameter to keep track of occurences, we can use return value to add it up
# Checking if list is not empty works really well for the recursive function, I thought I only needed it for first time input

find_target = 4

def recursiveFind(intlist):
    # checks if list is empty
    # found out you could do "if not intlist:" instead of "if len(intlist) == 0:", does same thing but cleaner code
    if not intlist:
        print("List is empty.")
        return 0
    # checks to see if first element matches the target
    if intlist[0] == find_target:
        # print(f"Found {find_target}!")
        
        # found out we can do this instead of creating a new parameter that keeps track of occurences
        # this adds 1 to the return value and calls the next recursive function that excludes the checked element
        return 1 + recursiveFind(intlist[1:])
    else:
        return recursiveFind(intlist[1:])
    

# Example case
inputList = [1, 2, 3, 4, 5, 6, 4, 4, 1, 4]
total_occurences = recursiveFind(inputList)
print(f"Total occurences of {find_target}: {total_occurences}")

````
